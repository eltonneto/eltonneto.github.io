```python
import pandas as pd
from sqlalchemy import create_engine

# Extrair dados do arquivo CSV de exemplo
df = pd.read_csv('caminho_para_o_arquivo.csv')

# Realizar transformações nos dados (você pode adicionar suas próprias transformações aqui)
df['nova_coluna'] = df['coluna_existente'].apply(lambda x: x*2)

# Carregar os dados transformados em um banco de dados (SQLite, por exemplo)
engine = create_engine('sqlite:///banco_de_dados.db')
df.to_sql('tabela', con=engine, if_exists='replace', index=False)

print("Processo ETL concluído com sucesso!")
```
